```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://www.yelp.com/biz/tesla-san-jose')
```

```python
r.status_code
```

```python
r.text
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
#type(soup.findAll(class_='fullwidth testiminial-info')) -> bs4.element.ResultSet
divs = soup.findAll(class_='comment__09f24__D0cxf css-qgunke')
```

```python
reviews = []
for div in divs:
    reviews.append(div.find("span").text)
```

```python
reviews
```

## Analysing the Data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.array(reviews), columns=['review'])
```

```python
df.head()
```

```python
len(df['review'])
```

```python
# df['review'].apply(lambda x: len(x.split()))

df['word_count'] = df['review'].apply(lambda x: len(x.split()))
```

```python
df.head()
```

```python
# df['review'].apply(lambda x: len(x))

df['char_count'] = df['review'].apply(lambda x: len(x))
```

```python
df.head()
```

```python
def average_words(x):
    words = x.split()
    return sum(len(word) for word in words) / len(words)
```

```python
# df['review'].apply(lambda x: average_words(x))


df['average_word_length'] = df['review'].apply(lambda x: average_words(x))
```

```python
df.head()
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
stop_words = stopwords.words('english')
```

```python
stop_words
```

```python
len(stop_words)
```

```python
# df['review'].apply(lambda x: len([word for word in x.split() if word.lower() in stop_words]))

df['stopwords_count'] = df['review'].apply(lambda x: len([word for word in x.split() if word.lower() in stop_words]))
```

```python
df.head()
```

```python
df['stopwords_rate'] = df['stopwords_count'] / df['word_count']
```

```python
df.head()
```

```python
df.sort_values(by='stopwords_rate')
```

```python
df.describe()
```

## Data Cleaning

```python
df['review']
```

```python
df['lowercase'] = df['review'].apply(lambda x: " ".join(word.lower() for word in x.split()))
```

```python
df.head()
```

```python
df['punctuation'] = df['lowercase'].str.replace('[^\w\s]', '')
```

```python
df.head()
```

```python
df['stopwords'] = df['punctuation'].apply(lambda x: " ".join(word for word in x.split() if word not in stop_words))
```

```python
df.head()
```

```python
pd.Series(" ".join(df['stopwords']).split()).value_counts()
```

```python
other_stop_words = ['get', 'told', 'would', 'week', 'us', 'test', 'right', 'left', 'one', 'even', 'also', 'go', 'ahead', 'asked']
```

```python
len(other_stop_words)
```

```python
df['clean_reveiew'] = df['stopwords'].apply(lambda x: " ".join(word for word in x.split() if word not in other_stop_words))
```

```python
df.head()
```

```python
pd.Series(" ".join(df['cleanreveiew']).split()).value_counts()
```

```python
df['clean_review_word_count'] = df['clean_reveiew'].apply(lambda x: len(x.split()))
df['clean_rate'] = df['clean_review_word_count'] / df['word_count']
```

```python
df.head()
```

## Lemmatization

```python
# Import textblob
#!pip install textblob 
```

```python
from textblob import Word
```

```python
# nltk.download('wordnet')
```

```python
df['lemmatized'] = df['clean_reveiew'].apply(lambda x: " ".join(Word(word).lemmatize() for word in x.split()))
```

```python
df.head()
```

## Sentiment Analysis

```python
from textblob import TextBlob
```

```python
df['polarity'] = df['lemmatized'].apply(lambda x: TextBlob(x).sentiment[0])
```

```python
df['subjectivity'] = df['lemmatized'].apply(lambda x: TextBlob(x).sentiment[1])
```
